```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
```

```python
dataset.DESCR
```

```python
dataset.keys()
```

```python
dataset.data
```

```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

```python
df["Target"] = dataset.target
```

```python
df
```

```python
X = df.drop("Target",axis=1)
y = df["Target"]
```

```python
X,y
```

```python
X.shape
```

```python
y.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=42)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
from sklearn import tree
```

```python
plt.figure(figsize=(12,10))
tree.plot_tree(model)
plt.show()
```

```python
plt.figure(figsize=(12,10))
tree.plot_tree(model,filled=True)
plt.show()
```

```python
## post pruning :  1. visualaization
#                  2. ccpAlpha
```

```python
model = DecisionTreeClassifier(max_depth=3)
model.fit(X_train,y_train)
```

```python
plt.figure(figsize=(12,10))
tree.plot_tree(model,filled=True)
plt.show()
```

```python
##  using ccpalpha
##  Complexity parameter used for Minimal Cost-Complexity Pruning. 
##  The subtree with the largest cost complexity that is smaller than ccp_alpha will be chosen.
##  By default, no pruning is performed. See Minimal Cost-Complexity Pruning for details.
```

```python
model = DecisionTreeClassifier(ccp_alpha=0.2)
model.fit(X_train,y_train)
```

```python
plt.figure(figsize=(12,10))
tree.plot_tree(model,filled=True)
plt.show()
```

```python
model = DecisionTreeClassifier(ccp_alpha=0.01)
model.fit(X_train,y_train)
```

```python
plt.figure(figsize=(12,10))
tree.plot_tree(model,filled=True)
plt.show()
```

```python
## Cost Complexity pruning
## CCpAlpha
##  e = E + Alpha(T)
```

```python
y_pre = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(classification_report(y_test,y_pre))
```

```python
confusion_matrix(y_test,y_pre)
```

```python
accuracy_score(y_test,y_pre)
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
data = fetch_california_housing()
```

```python
data.keys()
```

```python
data.DESCR
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df.head(5)
```

```python
df["target"] = data.target
```

```python
df.head(5)
```

```python
X = df.drop("target", axis=1)
y= df["target"]
X.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=24)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor = DecisionTreeRegressor()
```

```python
regressor.fit(X_train,y_train)
```

```python
y_pre = regressor.predict(X_test)
```

```python
r2_score(y_test,y_pre)
```

```python
y_pre
```
